```python
import plotly.express as px
import pandas as pd
```

```python
# Using pandas to read html table
url = 'https://e-service.cwb.gov.tw/HistoryDataQuery/MonthDataController.do?command=viewMain&station=466910&stname=%25E9%259E%258D%25E9%2583%25A8&datepicker=2022-02&altitude=837.6m'
table = pd.read_html(url,encoding='utf-8')[1]
table
```

```python
# Using Do loop to read html tables
yyyymm = [d.strftime('%Y-%m') for d in pd.date_range('2020-01','2020-12',freq='MS')]
```

```python
dfs = []
for wd in yyyymm :
    url = 'https://e-service.cwb.gov.tw/HistoryDataQuery/MonthDataController.do?command=viewMain&station=466920&stname=%25E8%2587%25BA%25E5%258C%2597&datepicker='+wd+'&altitude=5.3m'
    dfs.append(pd.read_html(url,encoding='utf-8')[1].loc(axis=1)[:,:,['ObsTime','Temperature']])
df = pd.concat(dfs)
dfs.clear()
df
```

```python
time = pd.date_range(start='2020-01-01', periods=366)
df.index = time
df.temperature.plot()
```
